```python
import matplotlib.pyplot as plt

class TreeNode:
    def __init__(self, name):
        self.name = name
        self.children = []

    def add_child(self, child_node):
        self.children.append(child_node)

def build_tree(node, node_dict):
    for key, value in node_dict.items():
        if isinstance(value, dict):
            child_node = TreeNode(key)
            node.add_child(child_node)
            build_tree(child_node, value)
        else:
            parts = value.split(', ')
            parent_node = node
            for part in parts:
                if '/liga/1/' in part:
                    part = part.split('/liga/1/')[1]
                if part.startswith('Klasa'):
                    # Check if a "Klasa" node with the same name already exists under the same parent
                    existing_klasa = next((child for child in parent_node.children if child.name == part), None)
                    if existing_klasa:
                        parent_node = existing_klasa
                    else:
                        child_node = TreeNode(part)
                        parent_node.add_child(child_node)
                        parent_node = child_node
                elif part.startswith('grupa'):
                    # Check if a "grupa" node with the same name already exists under the same parent
                    existing_grupa = next((child for child in parent_node.children if child.name == part), None)
                    if existing_grupa:
                        parent_node = existing_grupa
                    else:
                        child_node = TreeNode(part)
                        parent_node.add_child(child_node)
                        parent_node = child_node
                else:
                    child_node = TreeNode(part)
                    parent_node.add_child(child_node)
                    parent_node = child_node





def plot_tree(node, ax, depth=0, pos=(0, 0), max_depth=4, h_spacing=300, v_spacing=50, base_font_size=12, first_level_font_size=10):
    x, y = pos
    if depth == 1:
        font_size = first_level_font_size
    else:
        font_size = base_font_size - depth * 1.5

    ax.text(x, y, node.name, ha='center', va='center', fontsize=font_size, color='blue', bbox=dict(facecolor='lightgrey', edgecolor='black', boxstyle='round,pad=0.5'))

    if node.children and depth < max_depth:
        y -= v_spacing
        if depth == 1:
            h_spacing *= 1.5
        else:
            h_spacing /= 1.5

        for child in node.children:
            x_child = x - (len(node.children) - 1) * h_spacing / 2 + node.children.index(child) * h_spacing
            ax.plot([x, x_child], [y + v_spacing, y], color='black', linewidth=1.5)
            plot_tree(child, ax, depth + 1, (x_child, y), max_depth, h_spacing, v_spacing, base_font_size, first_level_font_size)

def visualize_tree(comps_tree):
    fig, ax = plt.subplots(figsize=(12, 8))
    plot_tree(comps_tree, ax)
    ax.axis('off')
    plt.show()



comps_tree = TreeNode(_BASE_URL)
build_tree(comps_tree, _COMPS_DICT[0])

visualize_tree(comps_tree)

```
